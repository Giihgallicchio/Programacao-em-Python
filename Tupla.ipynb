```python
# DESAFIO 72
# Crie um programa que tenha uma tupla totalmente preenchida com
# uma contagem por extenso, de zero até vinte.
# Seu programa deverá ler um número pelo teclado (entre 0 e 20) e
# mostra-lo por extenso.

tupla=('zero','um','dois','três','quatro','cinco','seis','sete','oito','nove','dez','onze','doze','treze','quatorze','quinze','dezesseis','dezessete','dezoito','dezenove','vinte')
numero=int(input('Digite um número entre 0 e 20: '))
print(f'Você digitou o número {tupla[numero]}')
```

```python
# DESAFIO 73
# Crie uma tupla preenchida com os 20 primeiros colocados da Tabela do
# Campeonato Brasileiro Serie B de Futebol, na ordem de colocação. Depois
# mostre:
# A) Apenas os 5 primeiros colocados.
# B) Os últimos 4 colocados da tabela.
# C) Uma lista com os times em ordem alfabética.
# D) Em qual posição na tabela está o time do Santos.

serireB=("América-MG", "Goiás", "Avaí", "Santos",
         "Sport", "Ceará", "Operário-PR", "Mirassol", "Coritiba" ,
         "Vila Nova", "Novorizontino"
         ,"Amazonas", "Chapecoense" ,"Ponte Preta",
         "CRB" ,"Paysandu", "Botafogo-SP"
         ,"Ituano", "Brusque" ,"Guarani")
# A) Apenas os 5 primeiros colocados.
print('Os 5 primeiros colocados são')
for pos, i in enumerate(serireB):
    if pos < 5:
        print(pos+1, i)
# B) Os últimos 4 colocados da tabela.
print('\n\n\nOs 4 ultimos colocados são')
for i in range(16,20):
    print(i+1, serireB[i])
# C) Uma lista com os times em ordem alfabética.
listaOrdenada=sorted(serireB)
for pos, time in enumerate(serireB):
    print(pos+1, time)
# D) Em qual posição na tabela está o time do Santos.
print(f'\n\nO santos esta na {serireB.index('Santos')+1} posição')
```

```python
# DESAFIO 74
# Crie um programa que vai gerar cinco números aleatórios e colocar
# em uma tupla.
# Depois disso, mostre a listagem de números gerados e também
# indique o menor e o maior valor que estão na Tupla.
from random import randint

numeros=tuple(randint(0,10) for x in range(5))

print(f'o maior numero é o {max(numeros)} e o menor numero é o {min(numeros)}')
```

```python
# DESAFIO 75
# Desenvolva um programa que leia quatro valores pelo teclado e
# guarde-os em uma tupla. No final, mostre:
# A) Quantas vezes apareceu o valor 9.
# B) Em que posição foi digitado o primeiro valor 3.
# C) Quais foram o números pares.

numeros=tuple(int(input(f'Digite o {x+1}º numero: ')) for x in range(4))

# A) Quantas vezes apareceu o valor 9.
print(f'o numero 9 apareceu: {numeros.count(9)} vezes' if 9 in numeros else 'O numero 9 não aparece')
# B) Em que posição foi digitado o primeiro valor 3.
print(f'o numero 3 apareceu na {numeros.index(3)+1}º posição ' if 3 in numeros else 'O numero 3 não aparece')
# C) Quais foram o números pares.
for x in numeros:
    if x%2==0:
        print(x, end=',')
```
