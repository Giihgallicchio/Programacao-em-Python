```python
#Projeto Final!
#Jogo da forca com os times do Brasileirão com as séries A,B e C
#By Giovanna A. Galichio ツ

import random
import unicodedata

def remover_acentos(texto):
    return ''.join(c for c in unicodedata.normalize('NFD', texto) if unicodedata.category(c) != 'Mn')

print('Bem-vindo ao jogo da forca dos times do Brasileirão!')
print('Você tem 6 chances para acertar o nome do time!')
print('Boa sorte!\n')

while True:
    print('Escolha a série do Brasileirão que você deseja jogar:')
    menu = input('Menu de Opções:\n[1] Série A\n[2] Série B\n[3] Série C\n[4] Sair do programa\n\nDigite sua Opção: ')

    if menu == '1':
        times = ('São Paulo', 'Palmeiras', 'Fluminense', 'Bahia', 'Flamengo', 'Coritiba', 'Corinthians', 'Bragantino', 'Grêmio', 'Athletico-PR', 'Vitória', 'Mirassol', 'Santos', 'Chapecoense', 'Atlético-MG', 'Vasco', 'Botafogo', 'Cruzeiro', 'Remo', 'Internacional')

    elif menu == '2':
        times = ('Ponte Preta', 'Atlético-GO', 'América-MG', 'Athletic', 'Avaí', 'Botafogo SP', 'Ceará SC', 'CRB', 'Criciúma', 'Cuiabá', 'Juventude', 'Fortaleza', 'Goiás', 'Novorizontino', 'Londrina', 'Náutico', 'Operário', 'São Bernardo', 'Sport Club', 'Vila Nova')

    elif menu == '3':
        times = ('Inter de Limeira', 'Confiança', 'Amazonas FC', 'Anápolis', 'Itabaiana', 'Barra', 'Botafogo-PB', 'Brusque', 'Ferroviario', 'Figueirense', 'Floresta', 'Guarani', 'Ituano', 'Maranhão', 'Maringá', 'Paysandu', 'Santa Cruz', 'Caxias', 'Volta Redonda', 'Ypiranga')

    elif menu == '4':
        print('Saindo do Programa...\nObrigado por Jogar!\nBy Giovanna A. Galichio ツ')
        break

    else:
        print('Opção inválida. Tente novamente.')
        continue

    times = list(times)
    random.shuffle(times)

    palavra = times.pop().upper()
    palavra_sem_acento = remover_acentos(palavra)

    acertos = []
    erros = 0

    while erros < 6:

        palavra_exibida = ''

        for i in range(len(palavra)):

            if remover_acentos(palavra[i]) in acertos or palavra[i] == ' ' or palavra[i] == '-':
                palavra_exibida += palavra[i] + ' '

            else:
                palavra_exibida += '_ '

        print(f"\nPalavra: {palavra_exibida}")

        letra = input('Digite uma letra: ').upper()

        if letra in acertos:
            print(f'Você já tentou a letra {letra}.')
            continue

        if letra in palavra_sem_acento:
            print(f'A letra {letra} está na palavra!')
            acertos.append(letra)

        else:
            erros += 1
            print(f'A letra {letra} não está na palavra.')
            print(f'Você tem {7 - erros} tentativas restantes.')

        palavra_acertada = True

        for letra_palavra in palavra_sem_acento:

            if letra_palavra != ' ' and letra_palavra != '-' and letra_palavra not in acertos:
                palavra_acertada = False
                break

        if palavra_acertada:
            print(f'Parabéns! Você acertou a palavra: {palavra}')
            break

    else:
        print(f'Você perdeu! A palavra era: {palavra}')

    novamente = input('\nDeseja jogar novamente? (S/N): ').lower()

    if novamente != 's':
        print('Obrigado por jogar!\nBy Giovanna A. Galichio ツ')
        break
```
