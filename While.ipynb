```python
# DESAFIO 58
# Melhore o jogo do DESAFIO 28, onde o computador vai “pensar” em
# um número entre 0 a 10. Só que agora o jogador vai tentar adivinhar
# até acertar, mostrando no final quantos palpites foram necessários para
# vencer.

from random import randint
numPC=randint(0,10)
pontucao=0
while True:
    print(numPC)
    numJogador=int(input('Acerte o numero de 0 a 10:\n'))
    if numPC==numJogador:
        print(f'Acertou! depois de {pontucao} tentativas')
        break
    else:
        print('Errou')
        pontucao+=1
```

```python
# DESAFIO 59
# Crie um programa que leia dois valores e mostre um menu na tela:
# [1] Somar
# [2] Multiplicar
# [3] Maior
# [4] Novos Números
# [5] Sair do programa
# Seu programa deverá realizar a operação solicitada em cada caso.

while True:
    num=input("Digite o primeiro numero:\n")
    num2=input("Digite o primeiro numero:\n")
    if num.isnumeric() and num2.isnumeric():
        num=int(num)
        num2=int(num2)
        break
    else:
        print('Digite apenas numeros!!')
while True:
    menu=input('MENU\n\n[1] Somar\n[2] Multiplicar\n[3] Maior\n[4] Novos Números\n[5] Sair do programa\n\nDigite: ')
    match menu:
        case '1':
            print(f'A soma dos numeros é {num+num2}')
        case '2':
            print(f'A multiplicação é {num*num2}')
        case '3':
            print(f'O maior valor é {max(num,num2)}')
        case '4':
            while True:
                num = input("Digite o primeiro numero:\n")
                num2 = input("Digite o primeiro numero:\n")
                if num.isnumeric() and num2.isnumeric():
                    num = int(num)
                    num2 = int(num2)
                    break
                else:
                    print('Digite apenas numeros!!')
        case '5':
            break
        case _:
            print('Escolha apenas opções validas!!')
```

```python
# DESAFIO 63
# Escreva um programa que leia um número n inteiro qualquer e mostre
# na tela os n primeiros elementos de uma sequência de Fibonacci.
# Exemplo:
# 0 –> 1 –> 1 –> 2 –> 3 –> 5 –> 8
num=int(input('Primeiro termo: '))
seguinte=num+1
resp=0
i=int(input('Quantidade de termos'))
while i > 0:
    resp= num + seguinte
    print(f'{resp} = {num} + {seguinte}')
    num=seguinte
    seguinte=resp
    i-=1
```

```python
# DESAFIO 69
#
# Crie um programa que leia a idade e o sexo de varias pessoas. A cada
# pessoa cadastrada, o programa deverá perguntar se o usuário quer ou
# não continuar. No final, mostre:
import os
maioreIdade=0
homens_qtd=0
mulheres20=0
i=1
while True:
    idade=input(f'Idade da {i}º pessoa: ')
    if not idade.isnumeric():
        continue
    else:
        idade=int(idade)
    sexo=input('Digite M para mulher e H para Homem: ').lower()
    if sexo != 'm' and sexo !='h':
        continue
    # A) Quantas pessoas tem mais de 18 anos.
    if idade>18:
        maioreIdade+=1
    # B) Quantos homens foram cadastrados.
    if sexo=='h':
        homens_qtd+=1
    # C) Quantas mulheres tem menos de 20 anos.
    if sexo=='m' and idade < 20:
        mulheres20+=1
    i+=1
    sair=input('Deseja sair? (s/n)').lower()
    if sair=='s':
        break
    os.system('cls')
os.system('cls')
print(f'Quantas pessoas tem mais de 18 anos? {maioreIdade}')
print(f'Quantos homens foram cadastrados? {homens_qtd}')
print(f'Quantas mulheres tem menos de 20 anos? {mulheres20}')
```

```python
# DESAFIO 70
# Crie um programa que leia o nome e o preço de vários produtos. O
# programa deverá perguntar se o usuário vai continuar. No final,
# mostre:
total=0
valorBarato=float('inf')
nomeBarato=''
produstoQtd100=0
i=0
while i < 10:
    nomeProd=input('Digite o nome do produto: ').title().strip()
    valorProd=float(input('Digite o valor do produto: '))
    if i == 0:
        valorBarato=valorProd
        nomeBarato = nomeProd
    # A) Qual é o total gasto na compra.
    total=valorProd+total
    # B) Quantos produtos custam mais de R$ 100,00.
    if valorProd > 100:
        produstoQtd100+=1
    # C) Qual é o nome do produto mais barato.
    if valorProd < valorBarato:
        valorBarato = valorProd
        nomeBarato = nomeProd
    i+=1
    sair = input('Deseja sair? (s/n)').lower()
    if sair == 's':
        break
print(f'A quantidade de produtos acima de 100R$ é {produstoQtd100}')
print(f'O produto mais barato foi o {nomeBarato}')
print(f'O total da compra foi de {total:,.2f}R$')
```
